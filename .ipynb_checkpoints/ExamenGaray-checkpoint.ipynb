# Laboral 

```python
#PARAMETROS

#staticos
recuperadosPrediccion=9759
tCrecimiento=int(media)
simulacion=214
LABORATORIO=26


recuperados={}
desempleados={}
reinserccionA={}
reinserccionR={}
reinserccionH={}
reinserccionV={}
reinserccionRE={}

alimentaciones={}
retail={}
hoteles={}
viajes={}
restaurante={}

```

```python
class Laborar(object):

    def __init__(self, env, lab):
            self.env = env
            self.lab=lab
            self.lab = simpy.Resource(env, lab)
            
    def descansar(self, personas):
        yield self.env.timeout(random.randint(2,7))
        #print("Persona, con numero", personas, "termina la recuperacion  el dia:  ", int(self.env.now))
      
    def teletrabajo(self, personas):
        yield self.env.timeout(random.randint(5,14))
        #print("Persona, con numero", personas, "termina teletrabajo  el dia: ", int(self.env.now))
      
    def valoracion(self, personas):
        yield self.env.timeout(random.randint(3,7))
        #print("Persona, con numero", personas, "termina la valoracion  el dia: ", int(self.env.now))
        
    def laboratorios(self, personas):
        yield self.env.timeout(random.randint(5,15))
        #print("Persona, con numero", personas, "sale del laboratorio el dia : ", int(self.env.now))
            
    def trabajo(self, personas):
        yield self.env.timeout(random.randint(5,14))
        #print("Persona, con numero", personas, "termina la etapa de trabajo  el dia: ", int(self.env.now))
def llegada_persona(env, laborar, persona):
    llega=env.now
    puestos=random.randint(1,100)
    if(puestos<=70):
        tipoEmpleo=random.randint(1,5)
        if (tipoEmpleo==1):
            alimentaciones[env.now] = alimentaciones[env.now] + 1 if env.now in alimentaciones else  1
            laborar.descansar(persona)
            laborar.teletrabajo(persona)
            bandera=True
            with laborar.lab.request() as lab:
                dias_esperando = random.randint(5,15)
                requerimientolab = yield lab | env.timeout(dias_esperando)
                wait = env.now - llega
                if lab in requerimientolab:
                    yield env.process(laborar.laboratorios(persona))
                    estado = random.randint(1,100)
                    if (estado <= 70):
                        laborar.trabajo(persona)
                        bandera=False
                        recuperados[env.now] = recuperados[env.now] + 1 if env.now in recuperados else  1
                        reinserccionA[env.now] = reinserccionA[env.now] + 1 if env.now in reinserccionA else  1
                    else:
                        laborar.valoracion(persona)
                        bandera=True

                else:
                    alimentaciones[env.now] = alimentaciones[env.now] + 1 if env.now in alimentaciones else  1

        elif(tipoEmpleo==2):
            restaurante[env.now] = restaurante[env.now] + 1 if env.now in restaurante else  1
            
            laborar.descansar(persona)
            laborar.teletrabajo(persona)
            bandera=True
            with laborar.lab.request() as lab:
                dias_esperando = random.randint(5,15)
                requerimientolab = yield lab | env.timeout(dias_esperando)
                wait = env.now - llega
                if lab in requerimientolab:
                    yield env.process(laborar.laboratorios(persona))
                    estado = random.randint(1,100)
                    if (estado <= 70):
                        laborar.trabajo(persona)
                        bandera=False
                        recuperados[env.now] = recuperados[env.now] + 1 if env.now in recuperados else  1
                        reinserccionRE[env.now] = reinserccionRE[env.now] + 1 if env.now in reinserccionRE else  1

                    else:
                        laborar.valoracion(persona)
                        bandera=True

                else:
                    restaurante[env.now] = restaurante[env.now] + 1 if env.now in restaurante else  1

        elif(tipoEmpleo==3):
            retail[env.now] = retail[env.now] + 1 if env.now in retail else  1
                
            laborar.descansar(persona)
            laborar.teletrabajo(persona)
            bandera=True
            with laborar.lab.request() as lab:
                dias_esperando = random.randint(5,15)
                requerimientolab = yield lab | env.timeout(dias_esperando)
                wait = env.now - llega
                if lab in requerimientolab:
                    yield env.process(laborar.laboratorios(persona))
                    estado = random.randint(1,100)
                    if (estado <= 70):
                        laborar.trabajo(persona)
                        bandera=False
                        recuperados[env.now] = recuperados[env.now] + 1 if env.now in recuperados else  1
                        reinserccionR[env.now] = reinserccionR[env.now] + 1 if env.now in reinserccionR else  1
                    else:
                        laborar.valoracion(persona)
                        bandera=True

                else:
                    retail[env.now] = retail[env.now] + 1 if env.now in retail else  1

        elif(tipoEmpleo==4):
            hoteles[env.now] = hoteles[env.now] + 1 if env.now in hoteles else  1
                
            laborar.descansar(persona)
            laborar.teletrabajo(persona)
            bandera=True
            with laborar.lab.request() as lab:
                dias_esperando = random.randint(5,15)
                requerimientolab = yield lab | env.timeout(dias_esperando)
                wait = env.now - llega
                if lab in requerimientolab:
                    yield env.process(laborar.laboratorios(persona))
                    estado = random.randint(1,100)
                    if (estado <= 70):
                        laborar.trabajo(persona)
                        bandera=False
                        recuperados[env.now] = recuperados[env.now] + 1 if env.now in recuperados else  1
                        reinserccionH[env.now] = reinserccionH[env.now] + 1 if env.now in reinserccionH else  1
                    else:
                        laborar.valoracion(persona)
                        bandera=True

                else:
                    hoteles[env.now] = hoteles[env.now] + 1 if env.now in hoteles else  1

        elif(tipoEmpleo==5):
            viajes[env.now] = viajes[env.now] + 1 if env.now in viajes else  1
            
            laborar.descansar(persona)
            laborar.teletrabajo(persona)
            bandera=True
            with laborar.lab.request() as lab:
                dias_esperando = random.randint(5,15)
                requerimientolab = yield lab | env.timeout(dias_esperando)
                wait = env.now - llega
                if lab in requerimientolab:
                    yield env.process(laborar.laboratorios(persona))
                    estado = random.randint(1,100)
                    if (estado <= 70):
                        laborar.trabajo(persona)
                        bandera=False
                        recuperados[env.now] = recuperados[env.now] + 1 if env.now in recuperados else  1
                        reinserccionV[env.now] = reinserccionV[env.now] + 1 if env.now in reinserccionV else  1
                    else:
                        laborar.valoracion(persona)
                        bandera=True

                else:
                     viajes[env.now] = viajes[env.now] + 1 if env.now in viajes else  1

    else:
        desempleados[env.now] = desempleados[env.now] + 1 if env.now in desempleados else  1

        
def ejecutar(env, tasa_crecimiento, recuperados):
    laborar = Laborar(env,LABORATORIO)
    for i in range(recuperadosPrediccion):
        asignar_laborar(env, laborar, i)
    personas = recuperadosPrediccion
    while True:
        yield env.timeout(1)
        for i in range(tasa_crecimiento):
            personas += 1
            asignar_laborar(env, laborar, personas)
        

def asignar_laborar(env, laborar, personas):
    existente=1
    if (existente == 1):
        #print("Llega nueva persona", personas, " en el día ", env.now)
        env.process(llegada_persona(env, laborar, personas))

print("Trabajos")
env=simpy.Environment()
env.process(ejecutar(env,tCrecimiento,recuperadosPrediccion))
env.run(until=simulacion)
     
```

El modelo  Polinomial, se trata de un modelo lineal en el que se incorporan nuevos predictores elevando el valor de los casos ya existentes a diferentes potencias. por lo que un polinomio de cuarto grado podría capturar mejor la relación entre las variables.
Los pronósticos probabilísticos pueden aplicarse a muchos ámbitos, que van desde el pronóstico del tiempo hasta las apuestas deportivas, pero son especialmente útiles para la optimización de la cadena de suministro.
Desde mi punto de vista, el modelo que realiza la mejor predicción es probabilístico, ya que se ajusta más a lo que esta pasando en el país. 

### Modelo exponencial

La vista de los resultados no parece que se puedan dar conclusiones definitivas sobre el ajuste exponencial, ya que si bien los puntos estimados de la función parecen acompañar a la recta a lo largo de su trayectoria, no es menos cierto que aquéllos parecen seguir un patrón ligeramente curvilíneo siempre por debajo de ésta. Es notable tambien que tiene un crecimiente rápido con el número de contagiados. Para el cáculo de este modelo se ha tomado en cuenta los datos encontrados y actualizados.

### Modelo Polinomico

La regresión polinomial se ajusta a una relación no lineal entre el valor de x y la correspondiente media condicional de y, denotada E (y | x), lo cual se ha utilizado para describir la progresión de esta epidemia que denotan el crecimiento de personas contagiadas en el país. Realizando un calculo a a partir de los datos encontrados. Es necesario tomar en cuenta el grado del polinomio con fin de encontrar una gráfica que tenga mayor ajuste a la tasa de crecimiento.
